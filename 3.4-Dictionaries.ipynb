##### Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs.
Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
## Creating Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(empty_dict)
print(type(empty_dict))
li=list()
print(type(li))
```

```python
student={"name":"Krish","age":32,"grade":24}
print(student)
print(type(student))
```

```python
# Single key is always used
student={"name":"Krish","age":32,"name":24}
print(student)
```

```python
## accessing Dictionary Elements
student={"name":"Krish","age":32,"grade":'A'}
print(student)
```

```python
## Accessing Dictionary elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## Modifying Dicitonary Elements
## Dictionary are mutable,so you can add, update or delete elements
print(student)
```

```python
student["age"]=33  ##update value for the key
print(student)
student["address"]="India" ## added a new key and value
print(student)
```

```python
del student['grade'] ## delete key and value pair

print(student)
```

No, a list cannot be used as a key in a Python dictionary.  
Dictionary keys must be immutable and hashable types, like strings, numbers, or tuples (with only immutable elements).

Lists are mutable, so they cannot be dictionary keys.  
If you try, youâ€™ll get a `TypeError`.

```python
# Examples of different data types as dictionary keys
dict_str = {"name": "Alice"}           # String as key
dict_int = {1: "one", 2: "two"}        # Integer as key
dict_float = {3.14: "pi"}               # Float as key
dict_tuple = {(1, 2): "tuple key",(3,4):"tuple sec key value"}      # Tuple as key (must be immutable)
dict_bool = {True: "yes", False: "no"} # Boolean as key
print(dict_str)
print(dict_int)
print(dict_float)
print(dict_tuple)
print(dict_bool)

# The following will cause an error because lists are mutable and not hashable:
# dict_list = {[1, 2]: "list key"}  # Uncommenting this line will raise TypeError
```

```python
print(dict_tuple[(1, 2)])
# dict_tuple[0] this is the wrong way , need to pass the exact key ,this type of calling should done in nested lists or tuples
```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictionatries, keys,values,or items

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate over values
for value in student.values():
    print(value)
```

```python
## Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Disctionaries
students={
    "student1":{"name":"Krish","age":32},
    "student2":{"name":"Peter","age":35}
}
print(students)
```

```python
## Access nested dictionaries elementss
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comphrehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Condition dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples

## Use a dictionary to count the frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
