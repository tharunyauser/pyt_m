#### Default Parameters

```python

#default parameters are used when no argument is provided during the function call.
def greet(name="Guest"):  # 'name' is a parameter
    print(f"Hello {name} Welcome To the paradise")

greet()           # Output: Hello Guest Welcome To the paradise
greet("Krish")    # Output: Hello Krish Welcome To the paradise
# "Krish" is an argument
```

if we call greet() without an argument, it won't raise an any TypeError.
bcoz it uses the default value "Guest" for the name parameter

```python
## missing positional argument

def greet(name):
    print(f"Hello {name} Welcome To the paradise")

greet()
```

### diff b/w parameter and argument
- **Parameter:** in the function definition
- **argument:** in the function call.
- Parameters are the variables listed in a function’s definition.
- Arguments are the actual values you pass to the function when you call it.

### Variable Length Arguments
### Positional And Keywords arguments

-Variable length arguments let you pass any number of values to a function.
- Use *args for any number of positional arguments (packed as a tuple).
- Use **kwargs for any number of keyword arguments (packed as a dictionary).

```python


#good practice and mostly we use print_numbers(*args)
def print_numbers(*krish):
    """
    Prints all positional arguments passed to the function.

    Parameters:
    *krish: Variable length positional arguments.

    Example:
    print_numbers(1, 2, 3)
    Output:
    1
    2
    3
    """
    for number in krish:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"tharunya")   # here all these parametrs represent to positional argument *krish
```

```python
## Positional arguments
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"xxx")
clss=["tharunya","rakshitha","prathyusha","vineela"]
print_numbers(clss)
```

### Keywords Arguments

```python

## kw->keyword     ,args-> positional arguments
def print_details(**kwargs):
    for key,value in kwargs.items():        #dict_anme.items() returns the list of the tuples of key values [(),(),.......]
        print(f"{key}:{value}")
```

```python
print_details(name="Krish",age="32",country="India")      #here I provided the parameters during run time 
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional arument :{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Krish",name="Krish",age="32",country="India")
```

```python
print_details(1,2,3,4,"Krish",name="Krish",age="32",country="India",45)
```
