```python
# String palindrome
def Palindrome(str):
    n=len(str)
    for i in range(0,n/2):
        if str[i]!=str[n-1-i]:
            return False
        
        return True
    
```

```python
#factorial of a number using recursion
def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
    
print(fact(5))
```
