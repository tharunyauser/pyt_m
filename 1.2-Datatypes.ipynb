## DataTypes
#### 1. Definition:

- Data types are a classification of data which tell the compiler or interpreter how the programmer intends to use the data.
- They determine the type of operations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data.

#### 2. Importance of Data Types in Programming
Explanation:

- Data types ensure that data is stored in an efficient way.
- They help in performing correct operations on data.
- Proper use of data types can prevent errors and bugs in the program.

```python
ch = 'A'        # This is a string, not a char type
print(type(ch)) # Output: <class 'str'>
```



All text data in Python is represented as strings (`str`), whether it's a single character or many characters.

## To get the ASCII (or Unicode) value of a character:
int('A')   # ❌ Error: invalid literal for int()
ord('A')   # ✅ Correct: returns 65

#### int() only works for:
- Numbers in string form → int("123")
- Booleans → int(True)
- Not for characters like 'A'

```python
print(ord('A'))    # z=65+25 = 90
print(ord('a'))    # z=97+25 = 122
print(ord('0'))
print(ord(' '))
print(ord('\n'))
print(ord('A')+3)    # z=65+3 = 68
```

```python
# To convert ASCII back to character
# Use chr():

print(chr(65))   # 'A'
print(chr(90))   # 'Z'
print(chr(97))   # 'a'
print(chr(90))   # 'Z'
print(chr(122))  # 'z'
print(chr(48))   # '0'
print(chr(32))   # ' '
print(chr(10))   # '\n'
```
