## Bitwise, ternary operators

In 1 st year of engineering, menu driven program was developed to perform basic arithmetic operations like addition, subtraction, multiplication, and division based on user input.


```python
# Bitwise Operators
# These operators work on bits and perform bit by bit operations.

a = 5      # 0b0101
b = 3      # 0b0011

# Bitwise AND
and_result = a & b   # 0b0001 -> 1

# Bitwise OR
or_result = a | b    # 0b0111 -> 7

# Bitwise XOR
xor_result = a ^ b   # 0b0110 -> 6

# Bitwise NOT
not_result = ~a      # -(a+1) -> -6

# Bitwise left shift
left_shift = a << 1  # 0b1010 -> 10

# Bitwise right shift
right_shift = a >> 1 # 0b0010 -> 2

print("Bitwise AND:", and_result)
print("Bitwise OR:", or_result)
print("Bitwise XOR:", xor_result)
print("Bitwise NOT:", not_result)
print("Left Shift:", left_shift)
print("Right Shift:", right_shift)

```

## Identity Operators

Identity operators check whether two variables point to the **same object in memory**, not just whether they have the same value.

### The Two Identity Operators

- **`is`** - Returns `True` if both variables reference the same object
- **`is not`** - Returns `True` if variables reference different objects

### Key Difference: `is` vs `==`

- **`==`** compares **values** (are the contents equal?)
- **`is`** compares **identity** (are they the same object in memory?)




```python
a = 10
b = 20
c = a

print(a is not b)  # True - a and b are different objects
print(a is c)      # True - c points to the same object as a
```



## Important Examples



```python
# Small integers are cached by Python
x = 10
y = 10
print(x is y)  # True - Python reuses the same object for small integers

# Lists with same values but different objects
list1 = [1, 2, 3]
list2 = [1, 2, 3]
print(list1 == list2)  # True - same values
print(list1 is list2)  # False - different objects in memory

# Same object reference
list3 = list1
print(list3 is list1)  # True - same object
```

## Why `x is y` Returns `True` for Small Integers

### Python's Integer Caching (Interning)

Python **pre-creates and caches** small integer objects (typically from **-5 to 256**) for performance optimization.

### Step-by-Step Explanation



```python
# Small integers are cached by Python
x = 10
y = 10
print(x is y)  # True - Python reuses the same object for small integers
```



### What Happens in Memory:

1. **When you create `x = 10`:**
   - Python checks: "Do I already have an integer object for 10?"
   - Yes! Python has a pre-created object for 10 in memory
   - `x` points to this existing object

2. **When you create `y = 10`:**
   - Python checks again: "Do I already have an integer object for 10?"
   - Yes! It's the same one that `x` is using
   - `y` points to the **exact same object** as `x`

3. **Result:**
   - Both `x` and `y` point to the **same memory location**
   - So `x is y` returns `True`

## Visual Representation



```python
Memory Address: 0x12345
    ↑         ↑
    |         |
    x = 10   y = 10
    
Both variables point to the SAME object!
```



## Contrast with Large Integers



```python
# Large integers are NOT cached
x = 1000
y = 1000
print(x is y)  # False - Different objects created

# But values are still equal
print(x == y)  # True - Same value
```

```python
Memory Address: 0x12345    Memory Address: 0x67890
    ↑                          ↑
    |                          |
    x = 1000                  y = 1000
    
Different objects in memory!
```





## Key Takeaway

- **Small integers (-5 to 256)**: Python reuses the same object → `is` returns `True`
- **Large integers**: Python creates separate objects → `is` returns `False`
- **Always use `==`** to compare values, **use `is`** only for identity checks (like `is None`)



## When to Use

- Use **`is`** to check if a variable is `None`: `if x is None:`
- Use **`==`** for value comparison in most other cases
- Use **`is`** when you specifically need to check object identity
