#### Abstraction
Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC,abstractmethod

## Abstract base cclass
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car enginer started")

def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

car=Car()
operate_vehicle(car)

```

In your code, abstraction is involved in the following places:

1. The class `Vehicle` is an abstract base class. It uses the `ABC` module and has an abstract method `start_engine`. This means you cannot create a direct object of `Vehicle`, and subclasses must implement `start_engine`.

2. The method `start_engine` in `Vehicle` is marked with `@abstractmethod`. This hides the implementation details and only shows the required feature (the method signature) to the user.

3. The function `operate_vehicle(vehicle)` works with any subclass of `Vehicle` that implements `start_engine`, without knowing how `start_engine` is implemented. This hides the complexity and shows only the necessary interface.

Summary:
- Abstraction is used by defining an abstract base class (`Vehicle`) and abstract method (`start_engine`).
- It hides the details and enforces a contract for subclasses.
- The user interacts with the object through a simple interface, not worrying about the internal implementation.
