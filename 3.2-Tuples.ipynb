##### Introduction to Tuples
Explanation:

Tuples are ordered collections of items that are immutable.
They are similar to lists, but their immutability makes them different.


```python
## creating a tuple
empty_tuple=()   # this is an empty tuple
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()    #() not a tuple, it's a list () function call
print(type(lst))
tpl=tuple()
print(type(tpl))
```

### Immumtable Nature of tuples.
as tuples are immutable we can't modify the tuple.
it didn' have the attribute methods like 

->append() and pop()

->insert()  and remove()

Lists support many more methods, such as:

append(), extend(), insert(), remove(), pop(), clear(), sort(), reverse(), and more
Summary:

Methods supported by tuples: count(), index()
Methods supported by lists (not tuples): append(), extend(), insert(), remove(), pop(), clear(), sort(), reverse(), etc.
Tuples are immutable, so they do not support methods that modify their contents. Lists are mutable and support many methods for changing their contents.

```python
# won't support bcoz these methods will change the original list
temp_list=(5, 23, 67, 9, 1, 4)
print(temp_list.sort())
print(temp_list.reverse())
print(temp_list.clear())
```

```python
## Packing and Unpacking tuple
## packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
##unpacking a tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))

## access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

### Difference Between Tuple and List in Python
| Feature         | List                | Tuple                |
|-----------------|---------------------|----------------------|
| Mutability      | Mutable (can change)| Immutable (cannot change) |
| Syntax          | [1, 2, 3]           | (1, 2, 3)            |
| Methods         | Many (append, pop, sort, clear, etc.) | Only count, index |
| Performance     | Slower (more flexible) | Faster (less flexible) |
| Use Case        | When you need to modify data | When data should not change |
| Memory Usage    | More                 | Less                 |
| Can be a dict key | No                  | Yes                  |
| Example         | my_list = [1, 2, 3]  | my_tuple = (1, 2, 3) |

**Summary:**
- Use a list when you need to add, remove, or change items.
- Use a tuple when you want to protect data from changes and use as dictionary keys.
