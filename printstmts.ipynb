### string formatting with the str.format() method.

How it works:

The curly braces {} are placeholders for values.

Inside the braces, :.2f means:

: starts the format specification.

.2f formats the value as a floating-point number with 2 digits after the decimal point.


```python
name = "Alice"
score = 95.678
print("Hello, {}!".format(name))                # Hello, Alice!
print("Score: {:.1f}".format(score))            # Score: 95.7
print("Value: {:d}".format(42))                 # Value: 42 (integer)
print("Hex: {:x}".format(255))                  # Hex: ff (hexadecimal)
```

Summary:

Use {} as placeholders in the string.

Add formatting instructions inside {}.

Call .format() with the values to insert.

## F string

```python
bonus_amount = 123.4567
print(f"Bonus Amount: ${bonus_amount:.2f}")
```

Explanation:

The f before the string allows you to insert variables directly inside {}.

{bonus_amount:.2f} formats bonus_amount as a float with 2 decimal places.

: format specification

### end 
You can set end to any string:

" " → space

"" → no gap

"\n" → newline (default)

"\t" → tab

"---" → custom separator

```python
print("Hello", end=" >>> ")
print("World")
print("Hello", end="---")
print("World")
```

#### sep
print() with sep
The sep parameter controls what goes between multiple arguments in a single print() call. By default, sep=" " (a space).

```python
print("apple", "banana", "cherry", sep=" - ")
print("apple", "banana", "cherry", sep=" | ")
print("apple", "banana", "cherry", sep=", ")
```

```python
print("apple", "banana", "cherry", sep=", ", end=".\n")
#sep=", " → separates items with a comma and space
#end=".\n" → ends the line with a period and newline
print("done")
```

sep → controls separators between items in one print().

end → controls what happens after the print statement.
